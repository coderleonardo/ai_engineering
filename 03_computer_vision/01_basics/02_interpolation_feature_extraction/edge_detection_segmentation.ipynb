```python
import cv2
import imageio.v2 as imageio
import itertools
import matplotlib.gridspec as gridspec
import matplotlib.pyplot as plt
from skimage.feature import hog
from skimage import exposure
from PIL import Image
from tensorflow.keras.preprocessing.image import ImageDataGenerator, img_to_array, load_img

import warnings
warnings.filterwarnings('ignore')
```

```python
def bgr2gray(imagem):
    return cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)
```

```python
def show_image(title, image, pos):
    plt.subplot(1, 3, pos)
    plt.imshow(image, cmap="gray")
    plt.title(title)
    plt.axis("off")
```

```python
image = cv2.imread("images/horse.jpg")
image_gray = bgr2gray(image)
```

```python
image_edges = cv2.Canny(image_gray, 100, 200)

_, segemnted_image = cv2.threshold(image_gray, 128, 255, cv2.THRESH_BINARY)
```

```python
plt.figure(figsize = (18, 6))

show_image('Original', image_gray, 1)
show_image('Canny edge detection', image_edges, 2)
show_image('Threshold segmentation', segemnted_image, 3)

plt.show()
```
